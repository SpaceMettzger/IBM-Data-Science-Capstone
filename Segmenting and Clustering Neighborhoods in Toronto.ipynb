## Part 1

Importing urllib and downloading the html data from Wikipedia. <br>
Then importing BeautifulSoup and creating a table from that data:

```python
import urllib.request

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)

from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "lxml")
right_table=soup.find('table', class_='wikitable sortable')
```

Create three empty lists for the three columns 'PostalCodes', 'Boroughs' and 'Neighborhoods'<br>
From the BeautifulSoup table find the entries with the 'tr' tag, from those find the entries with the 'td' tag and assign them to the variable cells <br>
If there are three entries in the variable cells (one for PostalCodes, one for Boroughs and one for Neighborhoods), take the text, turn it into a string, strip the linebreak charachter and append them to the corresponding list:

```python
PostalCodes = []
Boroughs = []
Neighborhoods = []

for row in right_table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells) == 3:
        PostalCodes.append(str(cells[0].find(text=True)).strip("\n"))
        Boroughs.append(str(cells[1].find(text=True)).strip("\n"))
        Neighborhoods.append(str(cells[2].find(text=True)).strip("\n"))
```

import Pandas and create a dataframe with three columns. The Column names are 'PostalCode', 'Borough' and 'Neighborhood'<br>
Fill the columns of the dataframe with the entries from the list of PostalCodes, Boroughs and Neighborhoods <br>
Print the first five entries to verify the data:

```python
import pandas as pd
df=pd.DataFrame(PostalCodes,columns=['PostalCode'])
df['Borough']=Boroughs
df['Neighborhood']= Neighborhoods

df.head()
```

Loop through every row of the Dataframe and create a list of all the rows where the value for Borough is 'Not assigned'.<br>
Then remove all the rows from the DataFrame and print out the first five rows to verify the data

```python
i = 0
index_list = []

for element in df['Borough']:
    if str(df.loc[i, 'Borough']).strip() == 'Not assigned':
        index_list.append(i)
        i += 1
    else:
        i += 1

df_mod = df.drop(df.index[index_list])
df_mod.head()
```

Used the code from the prevoius lab to get a list of venues for the Neighborhoods in our DataFrame

```python
import requests
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT = 100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Function to sort venues in descending order:

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Create a new dataframe with the name of the neighborhood and adding a ranking of the most common venues by calling the function return_most_common_venues for each row in the toronto_grouped DataFrame:

```python
import numpy as np
num_top_venues = 10

columns = ['Neighborhood']
columns.append('1st Most Common Venue')
columns.append('2nd Most Common Venue')
columns.append('3rd Most Common Venue')
columns.append('4th Most Common Venue')
columns.append('5th Most Common Venue')
columns.append('6th Most Common Venue')
columns.append('7th Most Common Venue')
columns.append('8th Most Common Venue')
columns.append('9th Most Common Venue')
columns.append('10th Most Common Venue')

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

import KMeans to cluster the toronto_grouped DataFrame. <br>
Drop the Neighborhood column (otherwise KMeans will fail), create the KMeans model with 5 clusters and fit it. <br>
Afterwards, return the created labels:

```python
from sklearn.cluster import KMeans
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

kmeans.labels_[0:] 
```

Insert the cluster labels into the neighborhood_venues_sorted dataframe and create a new dataframe in which the Dataframe from part 2 (with columns PostalCode,	Borough, Neighborhood, Latitude, Longitude) and the dataframe neighborhoods_venues_sorted are merged. <br>
Then drop all rows without values in the 'Cluster Labels' cloumn and display the first five rows of the new dataframe

```python
try:
    neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
except:
    pass

toronto_merged = df_complete

toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
toronto_merged.dropna(0, inplace = True)
toronto_merged.head()
```

Import folium to display the datapoints from the toronto_merged dataframe on a map of Toronto:

```python
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[43.7000, -79.4000], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
